## Descripción

El curso presenta al estudiante los aspectos de modelaje y diseño
algorítmico de un primer curso de optimización en redes. El curso tiene
un enfoque de programación matemática, como continuación natural de un
curso en optimización lineal (Principios de Optimización). Algunos temas
siguen un enfoque combinatorio, sirviendo como un primer curso en el
área de optimización combinatoria. El tema de optimización en redes
tiene una amplia aplicabilidad, además de encontrar relaciones directas
con la ciencia de redes y la matemática discreta. Algunos ejemplos de su
uso son: redes de transporte, redes de comunicaciones, planeación de
capital, programación de proyectos, mantenimiento y reemplazo de
equipos, programación de producción, entre otros.


## Objetivos

- Aplicar diversas técnicas de optimización del flujo de red para resolver problemas prácticos.
- Integrar enfoques basados en datos para mejorar la resolución de problemas.
- Desarrollar habilidades prácticas de modelado e implementación.

## Guía para las presentaciones científicas 
#### Instrucciones Generales:

Cada estudiante debe realizar dos presentaciones a lo largo del curso, cada una con una duración total de 10 minutos: 6 minutos de exposición y 4 minutos para preguntas y respuestas. Las presentaciones deben cubrir dos temas distintos:
1. **Presentación 1:** Un artículo donde sea propuesto un modelo de flujo en redes y preferiblemente relacionado a problemas de transporte y lógistica.
2. **Presentación 2:** Un artículo donde sea propuesto un modelo de flujo en redes y se propongan estrategias de aceleración para dichos modelos.

    *Información mas específica de las presentaciones se dará en las clases magistrales* 

#### Estructura Básica de la Presentación:

##### 1. Introducción (1 minuto)
Introducir brevemente el tema del artículo y su relevancia en el contexto de la investigación científica.
- **Contenido:**
  - Título del artículo y nombres de los autores.
  - Contexto general y motivación del estudio.
  - Objetivo principal del artículo.

##### 2. Contexto y Estado del Arte (1 minuto)
Proporcionar un contexto sobre la investigación previa relacionada con el artículo.
- **Contenido:**
  - Breve revisión de trabajos previos relevantes.
  - Cómo el artículo se posiciona respecto a investigaciones anteriores.
  - Justificación de la investigación.

##### 3. Descripción del Problema y Metodología (2 minutos)
Explicar el problema abordado en el artículo y la metodología utilizada.
- **Contenido:**
  - Descripción del problema de flujo en redes.
  - Formulación matemática y técnicas analíticas utilizadas.
  - Principales supuestos y simplificaciones.
  - Herramientas y datos utilizados.

##### 4. Resultados y Discusión (1.5 minutos)
Presentar y discutir los principales resultados del artículo.
- **Contenido:**
  - Resultados clave y cómo se obtuvieron.
  - Análisis de los resultados y comparación con trabajos previos.
  - Fortalezas y limitaciones de los resultados.

##### 5. Conclusiones y Futuras Líneas de Investigación (0.5 minuto)
Resumir las conclusiones del artículo y sugerir futuras líneas de investigación.
- **Contenido:**
  - Conclusiones principales del estudio.
  - Implicaciones prácticas y teóricas.
  - Posibles mejoras y futuras investigaciones sugeridas por los autores o por el presentador.

##### 6. Reflexión Crítica y Opinión Personal (1 minuto)
Proporcionar una reflexión crítica y personal sobre el artículo.
- **Contenido:**
  - Evaluación de la calidad de la investigación.
  - Opinión personal sobre la relevancia y el impacto del estudio.
  - Sugerencias para mejorar el estudio o aplicar sus resultados en otras áreas.

#### Recomendaciones para la Presentación:
- La dificultad del problema presentado, la calidad del periodico usado y la relación con el programa serán considerados en la evaluación.
- Leer el artículo varias veces para una comprensión profunda.
- Crear diapositivas claras y concisas, con gráficos y diagramas que apoyen la explicación.
- Ensayar la presentación para ajustarse al tiempo asignado y mejorar la fluidez.
- Ser claro y directo en la explicación, evitando tecnicismos innecesarios.
- Estar preparado para las preguntas.
- **Esta es una guía y como tal no responde a los componentes de una rúbrica**
- **Las secciones y los tiempos expuestos son apenas ilustrativos**
